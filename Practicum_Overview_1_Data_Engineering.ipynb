**5. Introduction to Data Analytics and Visualization**

Tujuan Pembelajaran:

* Memahami dasar-dasar analitik data dan visualisasi.
* Menggunakan perpustakaan Python untuk membuat visualisasi data.

Teori Singkat: Analitik data adalah proses mengumpulkan, memproses, dan menganalisis data untuk memperoleh wawasan yang dapat ditindaklanjuti. Visualisasi data membantu dalam mempresentasikan data dengan cara yang mudah dipahami.

Output yang Diharapkan:

* Grafik batang yang menunjukkan distribusi kategori dari kolom tertentu.
* Grafik sebar yang menunjukkan hubungan antara dua kolom.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data, trying different encodings if necessary
try:
    df = pd.read_csv('/content/booking_hotel.csv', encoding='utf-8')
except UnicodeDecodeError:
    df = pd.read_csv('/content/booking_hotel.csv', encoding='latin-1')

# Display the first few rows of the DataFrame to check the data
print(df.head())

# Print available columns in the DataFrame
print(df.columns)

# Choose a valid column name based on the printed column names
# Replace 'Hotel Name' with an actual column name from your DataFrame
valid_column_name = 'Hotel Name'  # Ensure this matches exactly as it appears in df.columns

if valid_column_name in df.columns:
    df[valid_column_name].value_counts().plot(kind='bar')
    plt.xlabel('Categories')
    plt.ylabel('Counts')
    plt.title('Sample Bar Plot')
    plt.show()
else:
    print(f"Column {valid_column_name} not found in DataFrame")

# For scatter plot, use valid column names
column_x = 'average_score'  # Replace with actual column name
column_y = 'total_number_of_reviews'  # Replace with actual column name

if column_x in df.columns and column_y in df.columns:
    plt.scatter(df[column_x], df[column_y])
    plt.xlabel('Average Score')
    plt.ylabel('Total Number of Reviews')
    plt.title('Scatter Plot')
    plt.show()
else:
    print(f"Columns {column_x} and/or {column_y} not found in DataFrame")
```

Melihat hasil plot yang sangat ramai dan tidak terbaca, sepertinya kolom yang kita pilih memiliki terlalu banyak kategori unik (seperti nama hotel yang berbeda-beda).

Untuk membuat grafik batang yang lebih informatif, kita bisa menampilkan hanya beberapa kategori teratas berdasarkan jumlah kemunculannya.

**Berikut adalah langkah-langkah untuk memperbaiki plot:**

* Filter Kategori Teratas: Tampilkan hanya beberapa kategori teratas berdasarkan jumlah kemunculannya. Misalnya, kita bisa menampilkan hanya 10 hotel teratas dengan jumlah pemesanan terbanyak.
* Mengatur Ukuran Plot: Atur ukuran plot agar lebih mudah dibaca.

Dengan menggunakan kode ini, plot yang dihasilkan akan jauh lebih rapi dan informatif, menampilkan hanya 10 hotel teratas berdasarkan jumlah pemesanan.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data, trying different encodings if necessary
try:
    df = pd.read_csv('/content/booking_hotel.csv', encoding='utf-8')
except UnicodeDecodeError:
    df = pd.read_csv('/content/booking_hotel.csv', encoding='latin-1')

# Check the available columns in your DataFrame
print(df.columns)

# Choose a valid column name based on the printed column names
valid_column_name = 'Hotel Name'  # Ensure this matches exactly as it appears in df.columns

# Display top 10 most frequent categories
top_10 = df[valid_column_name].value_counts().nlargest(10)

# Plot the top 10 categories
plt.figure(figsize=(10, 6))
top_10.plot(kind='bar')
plt.xlabel('Hotel Name')
plt.ylabel('Counts')
plt.title('Top 10 Hotels by Number of Bookings')
plt.xticks(rotation=45)
plt.show()
```
